```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calibration(x,y):
    x=np.array(x)
    y=np.array(y)
    p=np.polyfit(x,y,1)
    y_fit = p[0]*x+p[1]

    max_err =np.sqrt(np.max( (y_fit-y)**2/y**2   ))*100

    plt.title(f"y={p[0]:0.3f}*x + {p[1]:0.2f}, max err: {max_err:0.1f}%")


    plt.plot(x,y,".")
    plt.plot(x,y_fit)
    plt.show()
```

```python
## With tick_period_us = 1
pulse_width_set    =[10,20,30,40,50, 60, 70]
pulse_width_measure=[17,34,51,68,85,102,120]
calibration(pulse_width_set,pulse_width_measure)
```

```python
## With tick_period_us = 5
pulse_width_set    =[5,10,20,40,75,125,250]
pulse_width_measure=[5,10,21,41,77,128,256]
calibration(pulse_width_set,pulse_width_measure)
```

```python
## With tick_period_us = 10
pulse_width_set    =[10,20,50,100,200,500]
pulse_width_measure=[10,21,51,102,205,512]
calibration(pulse_width_set,pulse_width_measure)
```
